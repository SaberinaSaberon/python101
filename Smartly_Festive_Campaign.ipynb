```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sheet1 = pd.read_excel('/content/Smartly.io _ Analyst Test Marina Bay Sands.xlsx',sheet_name = 'Sheet1')
mapping_table = pd.read_excel('/content/Smartly.io _ Analyst Test Marina Bay Sands.xlsx', sheet_name = 'Mapping Table')

sheet1.head()
```

```python
sheet1.info()
```

```python
# check which row has total conversions greater than impressions, and should ask stakeholder if this is possible
sheet1['total_impressions'] = np.where(sheet1['Total Conversions']>sheet1['Impressions'],1,0)
sheet1.head()
```

```python
# for the meantime only filter rows with conversions less than total impressions
df = sheet1[sheet1['total_impressions']==0]
df
```

```python
# merge table with variationHash as unique key
df1 = pd.merge(df,mapping_table,on='variationHash',how='left')
df1.head()
```

```python
df1.nunique()
```

```python
df1.drop_duplicates(inplace=True)
df1.info()
```
